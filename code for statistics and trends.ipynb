```python
# Cell 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Cell 2: Load the dataset
df = pd.read_csv("data (9).csv")
df.head()
```

```python
# Cell 3: Show describe() and correlation matrix
print("Descriptive Statistics:")
print(df.describe())

print("\nCorrelation Matrix (numeric columns only):")
numeric_df = df.select_dtypes(include=[np.number])
print(numeric_df.corr())

```

```python
# Cell 4: Function to create a categorical bar chart

def plot_categorical(df):
    """
    Creates and saves a histogram of out-of-state tuition fees.
    """
    plt.figure(figsize=(7, 5))
    plt.hist(df['outstate'], bins=20, edgecolor='black')
    plt.title('Distribution of Out-of-State Tuition')
    plt.xlabel('Out-of-State Tuition ($)')
    plt.ylabel('Number of Colleges')
    plt.tight_layout()
    plt.savefig("categorical_plot_histogram.png")
    plt.show()

plot_categorical(df)
```

```python
# Cell 5: Function to create a relational scatter plot
def plot_relational(df):
    """
    Creates and saves a scatter plot of number of students accepted vs enrolled.
    """
    plt.figure(figsize=(7, 5))
    plt.scatter(df['accept'], df['enroll'], alpha=0.6)
    plt.title('Accepted vs Enrolled Students')
    plt.xlabel('Accepted')
    plt.ylabel('Enrolled')
    plt.tight_layout()
    plt.savefig("relational_plot.png")
    plt.show()
plot_relational(df)
```

```python
# Cell 6: Function to create a correlation heatmap
def plot_statistical(df):
    """
    Creates and saves a heatmap of correlation between numerical features.
    """
    numeric_df = df.select_dtypes(include=[np.number])  # Keep only numeric columns
    plt.figure(figsize=(12, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.savefig("statistical_plot.png")
    plt.show()

plot_statistical(df)
```
